```python
import requests as rq
import json
import datetime
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import time as time
import csv
import os

from datetime import datetime

os.system('touch earthquake_data2.csv')
```

```python
all_data = []
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2012-01-01&endtime=2017-01-01&minmagnitude=4.55&maxmagnitude=4.8'
earthquake_data = rq.get(url).json()

for i in range(0,len(earthquake_data['features'])):
    all_data.append({'Location':earthquake_data['features'][i]['properties']['place'], 'Magnitude':earthquake_data['features'][i]['properties']['mag'],'UTC_Time':earthquake_data['features'][i]['properties']['time'],'Longitude':earthquake_data['features'][i]['geometry']['coordinates'][0],'Latitude':earthquake_data['features'][i]['geometry']['coordinates'][1], 'Depth':earthquake_data['features'][i]['geometry']['coordinates'][2]})

print(all_data)
```

```python
data_df = pd.DataFrame(all_data)
data_df.head()
```

```python
data_df['Date'] = ''
data_df['Date_Time'] = '' 
data_df.head()
```

```python
len(data_df['Location'])
```

```python
def convert_to_date(string):
    date = datetime.fromtimestamp(string/1000)
    date_data = datetime.date(date)
    date_data = str(date_data).replace("-","")
    return (date_data)

def convert_to_date_time(string):
    date = datetime.fromtimestamp(string/1000)
    return (date)
```

```python
for index, row in data_df.iterrows():
    data_df['Date_Time'] = data_df['UTC_Time'].apply(convert_to_date_time)
    data_df['Date'] = data_df['UTC_Time'].apply(convert_to_date)
```
